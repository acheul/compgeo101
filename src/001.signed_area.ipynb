```python
from comp_geo_utils import signed_area, get_sign, area
import matplotlib.pyplot as plt
```

```python
data = [-5,-5, 5,0, 0,5]
print("signed_area:", signed_area(data))
print("area:", area(*data))

fig,ax = plt.subplots(1,1, figsize=(3,3))
j = len(data)-2
for i in range(0, len(data), 2):
  x1, y1 = data[j], data[j+1]
  x2, y2 = data[i], data[i+1]
  ax.arrow(x1,y1, (x2-x1),(y2-y1), head_width=.5)
  j = i
ax.set_title(get_sign(signed_area(data)))
plt.show()
```
